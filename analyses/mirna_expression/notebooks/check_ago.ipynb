```python
import pandas as pd
from pathlib import Path


PROJECTDIR = Path('.').resolve().parents[2]
lossfampath = f'{PROJECTDIR}/external_data/all_rodentlost_mirfams.list'
mirbase_mirgenedb_path = f'{PROJECTDIR}/external_data/mirgenedb_21_hsa_info.txt'
zhang_path = '../data/zhang2019_gastriccancer_agoclip.txt'

with open(lossfampath) as fh:
    lost_fams = [line.strip() for line in fh if line]
    print(lost_fams)

with open(zhang_path) as fh:
    zhang_mirnas = [line.strip() for line in fh if line.startswith('miR')]
            
    
print(zhang_mirnas)
    
```

```python
def mir2fam(mir):
    fam = '-'.join(mir.split('-')[1:3])
    return fam

col = []
with open(mirbase_mirgenedb_path) as fh:
    for line in fh:
        # print(line)
        mirgenedb_id, mirbase_id = line.split()[:2]
        mirgene_fam = mir2fam(mirgenedb_id)
        mirbase_mod = mirbase_id.replace('hsa-mir', 'miR')
        
        if mirgene_fam in lost_fams:
            col.append([mirgene_fam, mirgenedb_id, mirbase_mod])
            
df = pd.DataFrame(col, columns=['mirgene_fam', 'mirgenedb_id', 'mirbase_id'])
display(df)
```

```python
for mirfam in df.mirgene_fam.unique():
    mirfamdf = df[df.mirgene_fam == mirfam]
    print(mirfam, ', '.join(mirfamdf.mirbase_id.to_list()))
```

```python
def prune_zhang_mirnas(zhang_mirnas):
    col = []
    for mirid in zhang_mirnas:
        col.append(mirid.replace('-3p', '').replace('-5p', ''))
    return col

ago_mirnas = prune_zhang_mirnas(zhang_mirnas)
print(ago_mirnas)
```

```python
gastriccancer_lostmirna_ago = df[df.mirbase_id.isin(ago_mirnas)]
display(gastriccancer_lostmirna_ago)
```
