## How many genes that were lost in rodents are annotated as transcription factors

### Human Transcription factors database

* [Lambert SA, Jolma A, Campitelli LF, Das PK, Yin Y, Albu M, Chen X, Taipale J, Hughes TR, Weirauch MT.(2018) The Human Transcription Factors. Cell.](https://www.sciencedirect.com/science/article/pii/S0092867418301065?via%3Dihub)

```python
import pandas as pd

dbpath = '../data/DatabaseExtract_v_1.01.csv'

interestpath = '../results/symbol_filterlist_05.list'
with open(interestpath) as fh:
    genes_of_interest = [line.strip() for line in fh if line]

df = pd.read_csv(dbpath, sep=',', index_col=0)
df = df[df['Is TF?'] == 'Yes']
display(df)
print(df.columns)
```

```python
fildf = df[df['HGNC symbol'].isin(genes_of_interest)]
# fildf = df[df['HGNC symbol'].isin(genes_of_interest)]
display(fildf)
fildf.to_csv('../results/lost_TFs_HTFDB.tsv', sep='\t', index=False)
```

## Which binding domains?

```python
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns

motifcount = Counter(fildf.DBD)
num_motifs = len(motifcount)
labels = [f'{domain} ({motifcount[domain]})' for domain in motifcount.keys()]
colors = sns.color_palette('deep')[:num_motifs]
# print(motifcount)
plt.pie(motifcount.values(), labels=labels, colors=colors)
plt.title(f'Total transcription factors: {fildf.index.size} out of {len(genes_of_interest)}')
plt.savefig('tmp_tf_piechart.png', dpi=330)
plt.show()
```

## Are these more TFs than we would expect by chance?

M = total options

n = number positive outcomes

N = sample size

x = number of sucessfull draws

```python
from scipy.stats import hypergeom

M = 23823 # uniq gene ids in human protein set on ENSEMBL
n = df.index.size
N = len(genes_of_interest)
x = fildf.index.size

print(x, N, n, M)

pval = hypergeom.sf(x, M, n, N)
prb = hypergeom.pmf(x, M, n, N)


expected_draws = (N*n)/M

fold_enrichment = x / expected_draws

print(f'{fold_enrichment:.2f} enrichment with p-value {pval:.1e}')
```

## TFs and ZNFs

```python



tfs_of_interst = set(fildf['HGNC symbol'].values)
znf_of_interst = [gene for gene in genes_of_interest if gene.startswith('ZNF')]
znf_of_interst = set(znf_of_interst)

print(tfs_of_interst - znf_of_interst)
print(znf_of_interst - tfs_of_interst)

print(tfs_of_interst.union(znf_of_interst))
```

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2

set1 = len(tfs_of_interst - znf_of_interst)
set2 = len(znf_of_interst - tfs_of_interst)
intersection = len(tfs_of_interst.intersection(znf_of_interst))
union = len(tfs_of_interst.union(znf_of_interst))


venn2(subsets = (set1, set2, intersection), set_labels = ('TFs that are not ZNF', 'ZNFs not in RegNet'))
plt.title(f'Total transcription factors: {union} in set of {len(genes_of_interest)} genes')
plt.show()
```
