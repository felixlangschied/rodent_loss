## Load data

```python
import json
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from itertools import chain

theria_only = True
similarprots_rodentlossmirs = f'../data/allproteincoding_similarto_16lostfams.list'
similarprots_rodentlossmirs_json = f'../data/allproteincoding_similarto_16lostfams.json'
distmatdir = '/home/felixl/PycharmProjects/human_miRNA_evol/data/intermediate/distance_matrices'

pearsonpath = '../results/theria_only_pearson.tsv'
ppout = f'../results/symbol_simprot_pp.long'


def make_simprot_df(path, pearson):
    with open(similarprots_rodentlossmirs_json) as fh:
        simprots = json.load(fh)
        
    allprots = set()
    for tuplelist in simprots.values():
        for protid, corr in tuplelist:
        #print(protlist)
            allprots.add(protid)
            
    simprotdf = pd.DataFrame(index=list(allprots), columns=simprots.keys())
    allvaldf = pd.DataFrame(index=list(allprots), columns=simprots.keys())
    onedf = pd.DataFrame(index=list(allprots), columns=simprots.keys())
    for mirna, tuplelist in simprots.items():
        for protid, corr in tuplelist:
            simprotdf.loc[protid, mirna] = float(corr)
            onedf.loc[protid, mirna] = 1
            
            
    for mirna in allvaldf.columns.values:
        for protid in allprots:
            calcpearson = pearson.loc[mirna, protid]
            allvaldf.loc[protid, mirna] = calcpearson
            
    #display(simprotdf)
    return simprotdf, onedf, allvaldf


pdf = pd.read_csv(pearsonpath, sep='\t', index_col=0)
simprots, padf, allvals = make_simprot_df(similarprots_rodentlossmirs_json, pdf)
display(allvals)
display(simprots)
```

## Map Identifiers

* fDOG UniProt -> Gene-Symbol 
* Using QFO ID mapping file (http://www.ebi.ac.uk/reference_proteomes/)

```python
idmappath= '/home/felixl/PycharmProjects/human_miRNA_evol/data/mapping/UP000005640_9606.idmapping'

def read_idmap(path):
    m = {}
    with open(path) as fh:
        for line in fh:
            uniprot, typ, name = line.strip().split()
            if typ == 'Gene_Name':
                m[uniprot] = name
                #if not name in m:
                #    m[name] = []
                #m[name].append(uniprot)
        return m
            
            
idmap = read_idmap(idmappath)
```

## Make PhyloProfile input from Protein-PP and miRNA-PP

```python
import os
import glob

protphylo = '/share/gluster/Projects/felix/human_fdog_vertebrate/fdog_out/{0}/{0}.phyloprofile'
protpath = '/share/gluster/Projects/felix/human_fdog_vertebrate/human_vertebrata_orthologs_fdog.txt'
mirfampath = '/home/felixl/PycharmProjects/human_miRNA_evol/output/phyloprofile/mirgenedb2_0/family_seedpp.long'


alltax = set()
with open(protpath) as ph:
    for line in ph:
        rawid = line.split('|')[1].split('@')[1]
        alltax.add(f'ncbi{rawid}')


pp = []
with open(mirfampath, 'r') as mh:
    header = next(mh)
    for line in mh:
        group, taxid, gene, seddcon = line.strip().split()
        if not group in simprots.columns.values:
            continue
        if not taxid in alltax:
            continue

        pp.append(f'{line.strip()}\t1\n')

for protid in simprots.index.values:
    path = protphylo.format(protid)
    if not os.path.isfile(path):
        continue
    with open(path) as fh:
        header = next(fh)
        for line in fh:
            group, taxid, gene, fasf, fasb = line.strip().split()
            if group in idmap:
                symbol = idmap[group]
                ol = f'{symbol}\t{taxid}\t{symbol}\t{fasf}\t{fasb}\n'
            else:
                ol = line

            pp.append(ol)
            #pp.append(line)
            
            
with open(ppout, 'w') as of:
    of.write('geneID\tncbiID\torthoID\ts1\ts2\n')
    for l in reversed(pp):
        of.write(l)
print('# done')
```
